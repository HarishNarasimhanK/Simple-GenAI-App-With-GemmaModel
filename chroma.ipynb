```python
from langchain_chroma import Chroma 
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("alien.txt")
doc = loader.load()
doc
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 100,chunk_overlap = 20)
```

```python
doc = splitter.split_documents(doc)
doc
```

```python
embeddings = OllamaEmbeddings(model = "gemma2:2b")
vector_db = Chroma.from_documents(documents = doc , embedding= embeddings, persist_directory = "./chroma_db")
```

```python
vector_db
```

```python
query = "what is an alien life?"
ans = vector_db.similarity_search_with_score(query)
ans[0].page_content
```

```python
## laoding from chroma db
loaded_db = Chroma(persist_directory = "./chromadb",embedding_function = embeddings)
ans = loaded_db.similarity_search_with_score(query)
ans[0].page_content
```

```python
## Retriever options 
retriever = vector_db.as_retriever()
ans = retriever.invoke(query)
ans[0].page_content
```
